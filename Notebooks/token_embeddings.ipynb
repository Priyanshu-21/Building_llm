### Understanding how token embedding works under the hood
- Pre-trained model using `word2vec-google-news-300`.
- This model is trained on 300 features (dimensions) covering each word/ words semantic meaning.

```python
!pip install gensim
```

```python
import gensim.downloader as api

model = api.load('word2vec-google-news-300')

word_loader = model
```

```python
print("Getting similar word probablity details: ", word_loader.most_similar(['Computer']))
```

```python
# Get semantic meaning with math expression
# Expression Man + women - king ==> Semantic meaning resembles to Feminism
print(f'Get Semantic meanings : {word_loader.most_similar(positive= ['King', 'Women'], negative= ['Prince'], topn= 10)}')
```

```python
# Get different word probabilities
print(f'Similarity between Man and Women: {word_loader.similarity('Man', 'Women')}')

print(f'Similarity between Boy and Girl: {word_loader.similarity('Boy', 'Girl')}')

print(f'Similarity between Cricket and Dog: {word_loader.similarity('Cricket', 'Dog')}')
```

```python
import numpy as np
# Calculate Normalization between two different words in vector space

word_one = word_loader.get_vector('Boy')
word_two = word_loader.get_vector('Girl')

#Normalization
print(f'Normalization Difference: {np.linalg.norm(word_one) - np.linalg.norm(word_two)}')
```
